```python
import gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import string

# Sample text corpus
corpus = """
Word embeddings are a type of word representation that allows words to be represented as vectors in a continuous vector space.
The goal of word embeddings is to capture the semantic relationships between words.
For example, synonyms like 'happy' and 'joyful' should have similar vector representations.
The embeddings can be used for various downstream NLP tasks such as text classification, machine translation, etc.
"""

# Preprocessing: Tokenization and removing punctuation
def preprocess_text(text):
    # Tokenize words
    tokens = word_tokenize(text.lower())

    # Remove punctuation
    tokens = [word for word in tokens if word not in string.punctuation]
    return tokens

# Preprocess the corpus
processed_corpus = [preprocess_text(corpus)]

# --- CBOW Model (sg=0) ---
model_cbow = Word2Vec(sentences=processed_corpus, vector_size=100, window=5, min_count=1, sg=0)

# --- Skip-gram Model (sg=1) ---
model_skipgram = Word2Vec(sentences=processed_corpus, vector_size=100, window=5, min_count=1, sg=1)

# Check the vocabulary (words in the model)
print("Vocabulary in CBOW model:", list(model_cbow.wv.index_to_key))
print("Vocabulary in Skip-gram model:", list(model_skipgram.wv.index_to_key))

# Try words that are likely in the vocabulary (e.g., 'embeddings', 'word', 'semantic')
try:
    # Words similar to 'word' using CBOW
    similar_words_cbow = model_cbow.wv.most_similar('word', topn=5)
    print("CBOW - Words similar to 'word':", similar_words_cbow)

    # Words similar to 'word' using Skip-gram
    similar_words_skipgram = model_skipgram.wv.most_similar('word', topn=5)
    print("Skip-gram - Words similar to 'word':", similar_words_skipgram)

    # Similarity between 'word' and 'embeddings' for both models
    similarity_cbow = model_cbow.wv.similarity('word', 'embeddings')
    similarity_skipgram = model_skipgram.wv.similarity('word', 'embeddings')

    print("\nCBOW - Similarity between 'word' and 'embeddings':", similarity_cbow)
    print("Skip-gram - Similarity between 'word' and 'embeddings':", similarity_skipgram)

except KeyError as e:
    print(f"Word '{e.args[0]}' not found in vocabulary. Try other words.")
```
