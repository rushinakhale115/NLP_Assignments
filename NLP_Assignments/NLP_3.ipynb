```python
import nltk
import string
from collections import defaultdict, Counter
from nltk.util import ngrams
```

```python
nltk.download('punkt_tab')
```

```python
text = """
Paramagnetism is defined as the property of materials with atoms or ions that possess weakly interacting permanent magnetic moments, resulting in a positive magnetic susceptibility that varies with temperature and follows the Curie law.
"""

# Lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenization
words = nltk.word_tokenize(text)

print("Tokens:\n", words)
```

```python
N = 2

ngram_list = list(ngrams(words, N))

print(f"{N}-grams:\n", ngram_list)
```

```python
N = 3

ngram_list = list(ngrams(words, N))

print(f"{N}-grams:\n", ngram_list)
```

```python
model = defaultdict(Counter)

for gram in ngram_list:
    context = gram[:-1]
    next_word = gram[-1]
    model[context][next_word] += 1

print("Model:\n", dict(model))
```

```python
prob_model = {}

for context, counter in model.items():
    total = sum(counter.values())
    prob_model[context] = {word: count/total for word, count in counter.items()}

print("Probability Model:\n", prob_model)
```

```python
def predict_next_word(model, input_seq, N):
    input_tokens = input_seq.lower().split()
    context = tuple(input_tokens[-(N-1):])

    if context in model:
        return max(model[context], key=model[context].get)
    else:
        return "No prediction available"
```

```python
input_seq = "paramagnetism is"

predicted = predict_next_word(prob_model, input_seq, N)

print(f"Next word prediction: {predicted}")
```
